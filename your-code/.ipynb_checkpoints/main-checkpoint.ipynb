We will start by creating a standardized column for `CGPA`, `GRE Score` and `LOR`. We will name these columns `CGPA_std`, `GRE_std`, and `LOR_std` respecively.

Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
# your code here
###############################################One way############################################################
admission['CGPA_std']= (admission['CGPA'] - admission['CGPA'].mean()) /admission['CGPA'].std()
admission['GRE_std']= (admission['GRE Score'] - admission['GRE Score'].mean()) /admission['GRE Score'].std()
admission['LOR_std']= (admission['LOR'] - admission['LOR'].mean()) /admission['LOR'].std()


admission
```

```python
###############################################second way############################################################

admission['CGPA_std2'] = admission['CGPA'].apply(standardize,args=(np.mean(admission['CGPA']),np.std(admission['CGPA'])))
admission['GRE_std2'] = admission['GRE Score'].apply(standardize,args=(np.mean(admission['GRE Score']),np.std(admission['GRE Score'])))                                        
admission['LOR_std2'] = admission['LOR_std'].apply(standardize,args=(np.mean(admission['LOR_std']),np.std(admission['LOR_std'])))
admission
```

We will generate the decision choice at random using the code below. Please run the cell.

```python
# Libraries
from random import choices
```

```python
std_columns = ['CGPA_std', 'GRE_std', 'LOR_std']

decision_choice = choices(std_columns, k=admission.shape[0])
admission
```

Now create the deciding column using the `lookup` function. The lookup column is `decision_choice` found above. Call the column resulting from the lookup function `deciding_column` and add it to the `admissions` dataframe.

```python
# your code here
admission['deciding_column'] = admission.lookup(admission.index,decision_choice)
admission
```

Create a column called `decision` in the `admissions` dataframe. Assign 1 to this column if the value of `deciding_column` is greater than 0.8 and 0 otherwise.

```python
def take_decision(value):
    #Assign 1 to this column if the value of deciding_column is greater than 0.8 and 0 otherwise.
    if value > 0.8:
        return 1
    else:
        return 0
```

```python
# your code here
admission['decision'] = admission['deciding_column'].apply(take_decision)
admission
```

How many applicants will be accepted to the program using the decision column? Compute the result below.

```python
# your code here
admission['decision'].sum()
# 91 will be accepted
```
